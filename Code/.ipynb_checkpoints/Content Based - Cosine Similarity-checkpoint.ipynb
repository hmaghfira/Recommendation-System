```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn import preprocessing
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Dataset/film.csv')
df
```

```python
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
```

```python
categorical_col=df.genre #pake ini gausah mask
```

```python
df['genre_code'] =le.fit_transform(df.genre)
df['genre_code'].head()
```

```python
df
```

```python
# TRANSFORM WITH MIN MAX SCALLER
norm=df[['durasi', 'rating']]
min_max_scaler = preprocessing.MinMaxScaler()

x = 0
for n in norm :
    x=x+1
    np_x = min_max_scaler.fit_transform(norm[[n]])
    norm[n] = pd.DataFrame(np_x)
```

```python
df['durasi_minmax']=norm['durasi']
df['rating_minmax']=norm['rating']
```
